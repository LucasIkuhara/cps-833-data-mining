```python
import pandas as pd
import plotly.express as px
```

```python
# Ler dados de treino
df = pd.read_csv("data/train.csv")
df.head()
```

```python
# Checar se existem nulos ou faltantes
df.info()
```

```python
# Checar para erros de digitação em valores não numéricos
non_numerical = [
    "Gender",
    "family_history_with_overweight",
    "FAVC",
    "CAEC",
    "SMOKE",
    "SCC",
    "CALC",
    "MTRANS",
    "NObeyesdad",
]
for col in non_numerical:
    print(f"\nValores para {col}:")
    for v in df[col].unique():
        print(f" - {v}")
```

```python
from ipywidgets import widgets
from plotly.graph_objects import FigureWidget

graphs = []

for col in non_numerical:
    fig = px.histogram(df, x=col, title=f"Distribuição: {col}")
    graphs.append((fig))

widgets.HBox([graphs[0], graphs[1]])
```
